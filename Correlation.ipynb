```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_excel("top5-players24-25.xlsx")
df.head()
```

```python
# Diagrama de dispersiÃ³n
sns.scatterplot(data=df, x="Gls", y="Min")

# LÃ­nea de regresiÃ³n (opcional)
sns.regplot(data=df, x="Gls", y="Min", scatter=False, color="red")

# Personalizar el grÃ¡fico
plt.title("CorrelaciÃ³n entre minutos jugados y goles")
plt.ylabel("Minutos jugados")
plt.xlabel("Goles anotados")
plt.grid(True)
plt.show()
```

```python
topShooters = df.sort_values(by="Gls", ascending=False).head(10)

print(topShooters[["Pos","Player", "Squad", "Gls"]])  # Ajusta segÃºn el nombre real de las columnas
```

```python
# Crear figura
plt.figure(figsize=(10, 6))

# GrÃ¡fico de barras horizontales
plt.barh(topShooters["Player"], topShooters["Gls"], color="skyblue")

# Invertir el eje para que el que tiene mÃ¡s goles estÃ© arriba
plt.gca().invert_yaxis()

for i, (nombre, goles) in enumerate(zip(topShooters["Player"], topShooters["Gls"])):
    plt.text(goles + 0.5, i, str(goles), va="center")  

# TÃ­tulos y etiquetas
plt.title("Top 10 goleadores")
plt.xlabel("Nro de goles Temporada 24 - 25")
plt.ylabel("Nombre del Jugador")


# Mostrar grÃ¡fico
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Datos de edad
edades = df["Age"].dropna()  # AsegÃºrate de no tener NaN

# Definir los bins (los mismos que en tu grÃ¡fico, por ejemplo de 15 a 40 en pasos de 5)
bins = np.arange(15, 45, 5)  # 15-20, 20-25, ..., 40-45

# Histograma con numpy
frecuencias, bordes = np.histogram(edades, bins=bins)

# Mostrar los resultados
for i in range(len(frecuencias)):
    print(f"{bordes[i]}â€“{bordes[i+1]} aÃ±os: {frecuencias[i]} jugadores")
```

```python
# Datos de goles
goles = df["Gls"].dropna()  # AsegÃºrate de no tener NaN


bins = np.arange(0, 36, 5)  # 15-20, 20-25, ..., 40-45

# Histograma con numpy
frecuencias, bordes = np.histogram(goles, bins=bins)

# Mostrar los resultados
for i in range(len(frecuencias)):
    print(f"{bordes[i]}â€“{bordes[i+1]} goles: {frecuencias[i]} jugadores")
```

```python
# Histograma de Goles
plt.figure(figsize=(8, 5))
sns.histplot(df["Gls"], bins=5, kde=True, color="skyblue")
plt.title("DistribuciÃ³n de edades de los jugadores")
plt.xlabel("Goles")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Crear la figura
plt.figure(figsize=(8, 5))

# GrÃ¡fico de barras de conteo ordenado por frecuencia
ax = sns.countplot(data=df, x="Pos", order=df["Pos"].value_counts().index, palette="pastel")

# Agregar etiquetas a cada barra
for p in ax.patches:
    altura = p.get_height()
    ax.annotate(f'{int(altura)}',
                xy=(p.get_x() + p.get_width() / 2, altura),
                ha='center', va='bottom',
                fontsize=10)

# TÃ­tulos y ejes
plt.title("DistribuciÃ³n de jugadores por posiciÃ³n")
plt.xlabel("PosiciÃ³n")
plt.ylabel("Cantidad")
plt.tight_layout()
plt.show()
```

```python
goles_por_nacion = df.groupby("Nation")["Gls"].sum().sort_values(ascending=False).head(10)
print(goles_por_nacion)
```

```python
# Tomamos las top 10 nacionalidades con mÃ¡s goles
top_n = 10
top_n_goles = df.groupby("Nation")["Gls"].sum().sort_values(ascending=False).head(top_n)

# Crear grÃ¡fico de barras horizontales
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=top_n_goles.values, y=top_n_goles.index, palette="viridis")

# Agregar etiquetas al final de cada barra
for i, valor in enumerate(top_n_goles.values):
    plt.text(valor + 1, i, str(int(valor)), va='center')

# TÃ­tulos y etiquetas
plt.title(f"Top {top_n} nacionalidades con mÃ¡s goles")
plt.xlabel("Total de goles")
plt.ylabel("Nacionalidad")
plt.tight_layout()
plt.show()
```

```python
# Seleccionar solo columnas numÃ©ricas
numeric_cols = df.select_dtypes(include=['number']).columns

# Definir layout de la figura
cols = 4  # NÃºmero de grÃ¡ficos por fila
rows = int(np.ceil(len(numeric_cols) / cols))  # Calcula nÃºmero de filas necesarias

# Crear figura y ejes
fig, axes = plt.subplots(rows, cols, figsize=(5 * cols, 4 * rows))
axes = axes.flatten()  # Aplanar para indexar fÃ¡cilmente

# Generar los histogramas
for i, col in enumerate(numeric_cols):
    sns.histplot(data=df, x=col, kde=True, ax=axes[i])
    axes[i].set_title(col)

# Quitar subplots vacÃ­os si sobran
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
