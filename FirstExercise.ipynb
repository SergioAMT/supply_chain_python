```python
file_path = os.path.join(path, "top5-players24-25.xlsx")
df = pd.read_excel(file_path)
df.head(10)
```

```python
df.info()          # Tipos de datos, columnas nulas
df.describe()      # Estadísticas descriptivas
df.columns         # Nombres de columnas
```

```python
edad_media = df["Age"].mean()
edad_mediana = df["Age"].median()
edad_moda = df["Age"].mode()[0]

print("Edad - Media:", edad_media, "| Mediana:", edad_mediana, "| Moda:", edad_moda)
```

```python
matchplayed_media = df["MP"].mean()
matchplayed_mediana = df["MP"].median()
machtplaed_moda = df["MP"].mode()[0]

print("Edad - Media:", matchplayed_media, "| Mediana:", matchplayed_mediana, "| Moda:", machtplaed_moda)
```

```python
q1 = df["Gls"].quantile(0.25)
q2 = df["Gls"].quantile(0.50)  # Mediana
q3 = df["Gls"].quantile(0.75)

print("Goals - Q1:", q1, "| Q2:", q2, "| Q3:", q3)
```

```python
min_media = df["Min"].mean()
min_mediana = df["Min"].median()
min_moda = df["Min"].mode()[0]

print("min - Media:", min_media, "| Mediana:", min_mediana, "| Moda:", min_moda)
```

```python
varianza = df["Min"].var()
desviacion = df["Min"].std()
iqr = q3 - q1

print("Minutos jugados - Varianza:", varianza, "| Desviación estándar:", desviacion, "| IQR:", iqr)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(data=df, x="Min", y="Gls")
plt.title("Minutos jugados vs Goles")
plt.grid(True)
plt.show()
```
